```python
# Initial Setup:
# Import all the libraries we need

import matplotlib.pyplot as plt
import numpy as np
import csv
import math
import random

from BSE import market_session

# The next are helper functions that you will use later, if they don't make 
# much sense now, don't worry too much about it they will become clearer later:

# Use this to plot trades of a single experiment
def plot_trades(trial_id):
    prices_fname = trial_id + '_tape.csv'
    x = np.empty(0)
    y = np.empty(0)
    with open(prices_fname, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            time = float(row[1])
            price = float(row[2])
            x = np.append(x,time)
            y = np.append(y,price)

    plt.plot(x, y, 'x', color='black') 
    
# Use this to run an experiment n times and plot all trades
def n_runs_plot_trades(n, trial_id, start_time, end_time, traders_spec, order_sched):
    x = np.empty(0)
    y = np.empty(0)

    for i in range(n):
        trialId = trial_id + '_' + str(i)
        tdump = open(trialId + '_avg_balance.csv','w')

        market_session(trialId, start_time, end_time, traders_spec, order_sched, tdump, True, False)
        
        tdump.close()

        with open(trialId + '_tape.csv', newline='') as csvfile:
            reader = csv.reader(csvfile)
            for row in reader:
                time = float(row[1])
                price = float(row[2])
                x = np.append(x,time)
                y = np.append(y,price)

    plt.plot(x, y, 'x', color='black');

# !!! Don't use on it's own   
def getorderprice(i, sched, n, mode):
    pmin = min(sched[0][0], sched[0][1])
    pmax = max(sched[0][0], sched[0][1])
    prange = pmax - pmin
    stepsize = prange / (n - 1)
    halfstep = round(stepsize / 2.0)

    if mode == 'fixed':
        orderprice = pmin + int(i * stepsize)
    elif mode == 'jittered':
        orderprice = pmin + int(i * stepsize) + random.randint(-halfstep, halfstep)
    elif mode == 'random':
        if len(sched) > 1:
            # more than one schedule: choose one equiprobably
            s = random.randint(0, len(sched) - 1)
            pmin = min(sched[s][0], sched[s][1])
            pmax = max(sched[s][0], sched[s][1])
        orderprice = random.randint(pmin, pmax)
    return orderprice    

# !!! Don't use on it's own
def make_supply_demand_plot(bids, asks):
    # total volume up to current order
    volS = 0
    volB = 0

    fig, ax = plt.subplots()
    plt.ylabel('Price')
    plt.xlabel('Quantity')
    
    pr = 0
    for b in bids:
        if pr != 0:
            # vertical line
            ax.plot([volB,volB], [pr,b], 'r-')
        # horizontal lines
        line, = ax.plot([volB,volB+1], [b,b], 'r-')
        volB += 1
        pr = b
    if bids:
        line.set_label('Demand')
        
    pr = 0
    for s in asks:
        if pr != 0:
            # vertical line
            ax.plot([volS,volS], [pr,s], 'b-')
        # horizontal lines
        line, = ax.plot([volS,volS+1], [s,s], 'b-')
        volS += 1
        pr = s
    if asks:
        line.set_label('Supply')
        
    if bids or asks:
        plt.legend()
    plt.show()

# Use this to plot supply and demand curves from supply and demand ranges and stepmode
def plot_sup_dem(seller_num, sup_ranges, buyer_num, dem_ranges, stepmode):
    asks = []
    for s in range(seller_num):
        asks.append(getorderprice(s, sup_ranges, seller_num, stepmode))
    asks.sort()
    bids = []
    for b in range(buyer_num):
        bids.append(getorderprice(b, dem_ranges, buyer_num, stepmode))
    bids.sort()
    bids.reverse()
    
    make_supply_demand_plot(bids, asks) 

# plot sorted trades, useful is some situations - won't be used in this worksheet
def in_order_plot(trial_id):
    prices_fname = trial_id + '_tape.csv'
    y = np.empty(0)
    with open(prices_fname, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            price = float(row[2])
            y = np.append(y,price)
    y = np.sort(y)
    x = list(range(len(y)))

    plt.plot(x, y, 'x', color='black')   

# plot offset function
def plot_offset_fn(offset_fn, total_time_seconds):   
    x = list(range(total_time_seconds))
    offsets = []
    for i in range(total_time_seconds):
        offsets.append(offset_fn(i))
    plt.plot(x, offsets, 'x', color='black')  
```

```python
# First, configure the trader specifications
sellers_spec = [('PRDE', 30, {'k': 4, 'f_value':0.8, 's_min': -1.0, 's_max': +1.0}),]
buyers_spec = sellers_spec

traders_spec = {'sellers': sellers_spec, 'buyers': buyers_spec}

# Next, confiure the supply and demand (and plot it)
sup_range = (60,140)
dem_range = (60,140)

plot_sup_dem(10, [sup_range], 10, [dem_range], 'fixed')

# Next, configure order schedules
start_time = 0
hours=48
end_time = 60*60*hours
supply_schedule = [{'from': start_time, 'to': end_time, 'ranges': [sup_range], 'stepmode': 'fixed'}]
demand_schedule = [{'from': start_time, 'to': end_time, 'ranges': [dem_range], 'stepmode': 'fixed'}]

order_interval = 10
order_sched = {'sup': supply_schedule, 'dem': demand_schedule, 'interval': 10, 'timemode': 'drip-poisson'}

trial_id = 'prde'
tdump = open('prde_avg_balance.csv','w')
dump_all = True
verbose = True

# Now, run the market session
market_session(trial_id, start_time, end_time, traders_spec, order_sched, tdump, dump_all, verbose)

tdump.close()

# Finally, plot the trades that executed during the market session
plot_trades('prde')
```

# Running Homogenous Trials

```python
#trial that runs a session for all k (4,12) and for all f_value(0,2)
trail_id = 'all_k_to_all_f'
fname = trail_id + '_balances.csv'

tdump = open(fname, 'w')

for k in range(4,12):
    for f in np.arange(0.0,2.0,0.1):
        buyers_spec = [('PRDE', 30, {'k': k, 'f_value':f, 's_min': -1.0, 's_max': +1.0})] 
        sellers_spec = buyers_spec
        traders_spec = {'sellers': sellers_spec, 'buyers': buyers_spec}

        trial = 1
        while trial <= 1:
            trial_id = '%02d_%02f%s' % (k, f, trail_id)
            market_session(trial_id, start_time, end_time, traders_spec,
                           order_sched, tdump, False, True)
            tdump.flush()
            trial+=1

tdump.close()
```

```python
# trial where we keep k set to a specific number and evaluate all f (0,2)
# (k determined previously running expermienets on which k is the best for f_value = 0.8)

trail_id = 'set_k_to_all_f'
fname = trail_id + 'balances.csv'

tdump = open(fname, 'w')

for k in range(4):
    for f in np.arange(0.0,1.0,0.1):
        buyers_spec = [('PRDE', 30, {'k': k, 'f_value':f, 's_min': -1.0, 's_max': +1.0})] 
        sellers_spec = buyers_spec
        traders_spec = {'sellers': sellers_spec, 'buyers': buyers_spec}

        trial = 1
        while trial <= 1:
            trial_id = '%02d_%02f' % (k, f)
            market_session(trial_id, start_time, end_time, traders_spec,
                           order_sched, tdump, False, True)
            tdump.flush()
            trial+=1

tdump.close()
```

```python
#plotting the moving average of a single trail
df = pd.read_csv('prde_strats.csv',header=None,)

buyer_avg = []
seller_avg = []
total_avg = []

i=0
for row in range(0,len(df.index)):
    b_sum = 0
    s_sum = 0
    for col in range (8, 219 ,7):
        b_sum +=df.loc[row][col]
        
    for col in range (225, 428,7):
        s_sum +=df.loc[row][col]
    
    if(i!=0):
        buyer_avg.append(b_sum + buyer_avg[i-1])
        seller_avg.append(s_sum + seller_avg[i-1])
        total_avg.append(b_sum + s_sum + total_avg[i-1])
    else:
        buyer_avg.append(b_sum)
        seller_avg.append(s_sum)
        total_avg.append(b_sum + s_sum)
    i+=1
    
moving_avg_b = []
moving_avg_s = []
moving_avg_t = []

for i in range(1, len(buyer_avg)+1):
    moving_avg_b.append(buyer_avg[i-1]/i)
    moving_avg_s.append(seller_avg[i-1]/i)
    moving_avg_t.append(total_avg[i-1]/i)

t_period = np.linspace(0,hours,num=hours)
plt.plot(t_period, moving_avg_t, color='black',linestyle='dotted', marker='x') 
plt.plot(t_period, moving_avg_b, color='red', linestyle='dotted', marker='x') 
plt.plot(t_period, moving_avg_s, color='blue', linestyle='dotted', marker='x') 
plt.show()
```

```python
#calculating the moving average of pps for one trail - function

def moving_avg(trader_avg):
    trader_mov_avg = []
    
    for i in range(1, len(trader_avg)+1):
        trader_mov_avg.append(trader_avg[i-1]/i)
        
    return trader_mov_avg
    
def calculate_plot_ppc(file_name, hours):
    df = pd.read_csv(file_name+'_strats.csv',header=None)
    buyer_avg = []
    seller_avg = []
    
    i=0
    for row in range(0,len(df.index)):
        b_sum = 0
        s_sum = 0
        for col in range (8, 219 ,7):
            b_sum +=df.loc[row][col]

        for col in range (225, 428,7):
            s_sum +=df.loc[row][col]

        if(i!=0):
            buyer_avg.append(b_sum + buyer_avg[i-1])
            seller_avg.append(s_sum + seller_avg[i-1])
        else:
            buyer_avg.append(b_sum)
            seller_avg.append(s_sum)
        i+=1

    b_moving_avg = moving_avg(buyer_avg)
    s_moving_avg = moving_avg(seller_avg)
    t_moving_avg = np.add(b_moving_avg,s_moving_avg)

#     t_period = list(range(0,hours))
#     plt.plot(t_period, t_moving_avg, color='black',linestyle='dotted') 
#     plt.plot(t_period, b_moving_avg, color='red', linestyle='dotted') 
#     plt.plot(t_period, s_moving_avg, color='blue', linestyle='dotted') 
#     plt.show()
    return t_moving_avg
    
calculate_plot_ppc('prde_0', hours)
```

```python
#plotting the moving average of a multiple trails

def process_multiple_ppc(num_trails, hours):
    colors = ['black', 'pink', 'red', 'green', 'blue']
    all_trail_t_mov_avg = []
    
    for trial in list(range(1,num_trails)):
        trail_t_mov_avg = calculate_plot_ppc('prde_%d'%trial, hours)
        all_trail_t_mov_avg.append(trail_t_mov_avg)
        
        t_period = list(range(0,hours))
        plt.plot(t_period, trail_t_mov_avg, color=colors[trial],linestyle='dotted') 
        
    plt.show()
    return(all_trail_t_mov_avg)

all_trail_t_mov_avg = process_multiple_ppc(5, 48)
```

```python
#calculaes the average of each total moving avg and finds the index of the highest one

all_means = []
for i in range(len(all_trail_t_mov_avg)):

    mean_i = np.mean(all_trail_t_mov_avg[i])
    all_means.append(mean_i)

print(np.argmax(all_means), max(all_means))
```
